# 繪製函數圖表

1.   函數: 繪製蝴蝶>>需用到笛卡爾座標轉換
2.   線條特殊: 彩色漸變
3.   scatter plot



```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 蝴蝶曲線函數
theta = np.linspace(0, 12*np.pi, 1000)
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + np.sin(theta/12)**5

# 極座標與笛卡爾座標轉換
x = r * np.cos(theta)
y = r * np.sin(theta)

# 繪製彩色漸變曲線
plt.figure(figsize=(6,6))
scatter = plt.scatter(x, y, c=theta, cmap="rainbow", s=1.5)# c:顏色隨角度變化
plt.title("function of butterfly")
plt.axis("equal")
plt.colorbar(scatter, label="Theta")
plt.grid(True)

plt.show()
```
